```python
from pathlib import Path
import os
import re
import random
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
import torch
from torch.utils.data import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    DataCollatorWithPadding,
    Trainer,
    TrainingArguments
)
from transformers import EarlyStoppingCallback
import nltk
from nltk.stem import WordNetLemmatizer

# -------------------------
# NLTK Data Download with Error Handling
# -------------------------
def download_nltk_data():
    try:
        nltk.data.find('tokenizers/punkt')
    except LookupError:
        nltk.download('punkt', quiet=True)
    try:
        nltk.data.find('corpora/wordnet')
    except LookupError:
        nltk.download('wordnet', quiet=True)

download_nltk_data()

# -------------------------
# Config
# -------------------------
TRAIN_PATH = os.getenv("TRAIN_PATH", "/content/train.csv")
TEST_PATH = os.getenv("TEST_PATH", "/content/test.csv")
SAMPLE_SUB_PATH = os.getenv("SAMPLE_SUB_PATH", "/content/sample_submission.csv")
OUTPUT_DIR = os.getenv("OUTPUT_DIR", "/content/outputs")

MODEL_NAME = os.getenv("MODEL_NAME", "microsoft/deberta-v3-base")
EPOCHS = int(os.getenv("EPOCHS", "8"))
BATCH_SIZE = int(os.getenv("BATCH_SIZE", "16"))
LR = float(os.getenv("LR", "2e-5"))
SEED = int(os.getenv("SEED", "42"))
MAX_LENGTH = int(os.getenv("MAX_LENGTH", "512"))
N_FOLDS = int(os.getenv("N_FOLDS", "3"))

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

def set_seed(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

def clean_text(text: str) -> str:
    """Enhanced text cleaning pipeline"""
    if not isinstance(text, str) or not text.strip():
        return ""

    try:
        # Lowercase
        text = text.lower()
        # Remove special characters and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        # Tokenize and lemmatize
        words = nltk.word_tokenize(text)
        words = [lemmatizer.lemmatize(word) for word in words]
        # Remove extra whitespace
        return ' '.join(words).strip()
    except Exception as e:
        print(f"Error cleaning text: {e}")
        return text

def combine_text(row: pd.Series) -> str:
    """Improved text formatting with special tokens"""
    q = clean_text(row.get("QuestionText", ""))
    a = clean_text(row.get("MC_Answer", ""))
    e = clean_text(row.get("StudentExplanation", ""))

    parts = []
    if q: parts.append(f"[QUESTION] {q}")
    if a: parts.append(f"[ANSWER] {a}")
    if e: parts.append(f"[EXPLANATION] {e}")

    return " ".join(parts) if parts else ""

class TextDataset(Dataset):
    def __init__(self, tokenizer, texts, labels=None, max_length=384):
        self.tokenizer = tokenizer
        self.texts = texts
        self.labels = labels
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        item = self.tokenizer(
            self.texts[idx],
            truncation=True,
            max_length=self.max_length
        )
        if self.labels is not None:
            item["labels"] = int(self.labels[idx])
        return item

def map_at_3(y_true: np.ndarray, y_pred_top3: np.ndarray) -> float:
    correct = 0.0
    for t, preds in zip(y_true, y_pred_top3):
        if t in preds:
            rank = list(preds).index(t) + 1
            correct += 1.0 / rank
    return correct / len(y_true) if len(y_true) > 0 else 0.0

def compute_metrics(eval_pred):
    logits, labels_true = eval_pred
    top3 = np.argsort(-logits, axis=1)[:, :3]
    score = map_at_3(labels_true, top3)
    return {"map3": score}

def train_model(train_df, val_df, tokenizer, collator, params):
    """Train and evaluate a single model fold"""
    tr_texts = train_df["text"].tolist()
    tr_labels = train_df["label"].values
    va_texts = val_df["text"].tolist()
    va_labels = val_df["label"].values

    train_ds = TextDataset(tokenizer, tr_texts, tr_labels, max_length=MAX_LENGTH)
    val_ds = TextDataset(tokenizer, va_texts, va_labels, max_length=MAX_LENGTH)

    device_count = torch.cuda.device_count()
    per_device_batch = max(1, params["batch_size"] // max(1, device_count))

    model = AutoModelForSequenceClassification.from_pretrained(
        MODEL_NAME,
        num_labels=params["num_labels"],
        id2label=params["id2label"],
        label2id=params["label2id"]
    )

    args = TrainingArguments(
        output_dir=os.path.join(params["output_dir"], f"fold_{params['fold']}"),
        per_device_train_batch_size=per_device_batch,
        per_device_eval_batch_size=per_device_batch,
        num_train_epochs=params["epochs"],
        learning_rate=params["lr"],
        weight_decay=params["weight_decay"],
        eval_strategy="epoch",
        save_strategy="epoch",
        save_total_limit=1,
        load_best_model_at_end=True,
        metric_for_best_model="eval_map3",
        greater_is_better=True,
        fp16=torch.cuda.is_available(),
        logging_steps=50,
        report_to=[],
        seed=params["seed"],
        lr_scheduler_type="cosine",
        warmup_ratio=0.1,

    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=train_ds,
        eval_dataset=val_ds,
        tokenizer=tokenizer,
        data_collator=collator,
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=2, early_stopping_threshold=0.01)]
    )

    trainer.train()
    eval_results = trainer.evaluate()
    return eval_results["eval_map3"]

def main():
    set_seed(SEED)
    Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)

    # Load data
    train_df = pd.read_csv(TRAIN_PATH)
    test_df = pd.read_csv(TEST_PATH)
    sample_sub = pd.read_csv(SAMPLE_SUB_PATH)

    # Clean and preprocess text
    for c in ["QuestionText", "MC_Answer", "StudentExplanation"]:
        if c in train_df.columns:
            train_df[c] = train_df[c].fillna("")
        if c in test_df.columns:
            test_df[c] = test_df[c].fillna("")

    if "Category" not in train_df.columns or "Misconception" not in train_df.columns:
        raise ValueError("Train CSV must contain 'Category' and 'Misconception'.")

    label_mask = train_df["Category"].notna() & train_df["Misconception"].notna()
    train_df = train_df[label_mask].copy()
    train_df["target_str"] = train_df["Category"].astype(str) + ":" + train_df["Misconception"].astype(str)

    labels = sorted(train_df["target_str"].unique())
    label2id = {lbl: i for i, lbl in enumerate(labels)}
    id2label = {i: lbl for lbl, i in label2id.items()}
    train_df["label"] = train_df["target_str"].map(label2id).astype(int)

    # Apply enhanced text processing
    train_df["text"] = train_df.apply(combine_text, axis=1)
    test_df["text"] = test_df.apply(combine_text, axis=1)

    # Initialize tokenizer and collator
    tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)
    collator = DataCollatorWithPadding(tokenizer=tokenizer)

    # K-Fold Cross Validation
    kf = KFold(n_splits=N_FOLDS, shuffle=True, random_state=SEED)
    fold_scores = []

    print(f"\nStarting {N_FOLDS}-fold cross-validation...")
    for fold, (train_idx, val_idx) in enumerate(kf.split(train_df)):
        print(f"\nFold {fold + 1}/{N_FOLDS}")

        fold_train = train_df.iloc[train_idx]
        fold_val = train_df.iloc[val_idx]

        score = train_model(
            fold_train,
            fold_val,
            tokenizer,
            collator,
            params={
                "batch_size": BATCH_SIZE,
                "epochs": EPOCHS,
                "lr": LR,
                "weight_decay": 0.05,
                "num_labels": len(labels),
                "id2label": id2label,
                "label2id": label2id,
                "output_dir": OUTPUT_DIR,
                "fold": fold,
                "seed": SEED
            }
        )
        fold_scores.append(score)
        print(f"Fold {fold + 1} MAP@3: {score:.4f}")

    print(f"\nMean MAP@3 across folds: {np.mean(fold_scores):.4f}")

    # Train final model on full data
    print("\nTraining final model on full dataset...")
    full_ds = TextDataset(tokenizer, train_df["text"].tolist(), train_df["label"].values, max_length=MAX_LENGTH)

    device_count = torch.cuda.device_count()
    per_device_batch = max(1, BATCH_SIZE // max(1, device_count))

    full_args = TrainingArguments(
        output_dir=os.path.join(OUTPUT_DIR, "full"),
        per_device_train_batch_size=per_device_batch,
        per_device_eval_batch_size=per_device_batch,
        num_train_epochs=EPOCHS,
        learning_rate=LR,
        weight_decay=0.05,
        eval_strategy="no",
        save_strategy="epoch",
        save_total_limit=1,
        fp16=torch.cuda.is_available(),
        logging_steps=50,
        report_to=[],
        seed=SEED,
        lr_scheduler_type="cosine",
        warmup_ratio=0.1,
    )

    final_model = AutoModelForSequenceClassification.from_pretrained(
        MODEL_NAME,
        num_labels=len(labels),
        id2label=id2label,
        label2id=label2id
    )

    full_trainer = Trainer(
        model=final_model,
        args=full_args,
        train_dataset=full_ds,
        tokenizer=tokenizer,
        data_collator=collator
    )

    full_trainer.train()

    # Predict on test set
    test_ds = TextDataset(tokenizer, test_df["text"].tolist(), labels=None, max_length=MAX_LENGTH)
    logits = full_trainer.predict(test_ds).predictions
    top3 = np.argsort(-logits, axis=1)[:, :3]
    pred_strs = [" ".join(id2label[idx] for idx in row) for row in top3]

    # Prepare submission
    submission = sample_sub.copy()
    col = "Category:Misconception"
    if col not in submission.columns:
        submission[col] = ""
    submission[col] = pred_strs[:len(submission)]
    out_path = os.path.join(OUTPUT_DIR, "submission.csv")
    submission.to_csv(out_path, index=False)

    print("\nCross-validation scores:", fold_scores)
    print(f"Mean MAP@3: {np.mean(fold_scores):.4f}")
    print("Saved submission to:", out_path)
```

```python
if __name__ == "__main__":
    main()
```
